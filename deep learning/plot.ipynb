```python
import numpy as np
import math
import matplotlib.pyplot as plt

def plot_function(func, xmin, xmax, steps=1000, title='Function Plot'):
    """
    绘制给定函数的图形。
    
    参数:
    func: 要绘制的函数，作为一个接受单个变量的函数对象。
    xmin, xmax: 绘图的x轴范围。
    steps: 在x轴上的采样点数量。
    title: 图形的标题。
    """
    x = np.linspace(xmin, xmax, steps)  # 在xmin和xmax之间生成steps个等间距的点
    y = func(x)                          # 计算对应的y值
    
    plt.figure(figsize=(10, 6))          # 设置图形大小
    plt.plot(x, y, label=f'y = {func.__name__}(x)')  # 绘制函数曲线
    plt.title(title)                      # 设置图形标题
    plt.xlabel('x')                       # 设置x轴标签
    plt.ylabel('y')                       # 设置y轴标签
    plt.axhline(0, color='black', lw=0.5) # 绘制x轴
    plt.axvline(0, color='black', lw=0.5) # 绘制y轴
    plt.legend()                         # 显示图例
    plt.grid(True)                        # 显示网格
    plt.show()                           # 显示图形

# 定义你要绘制的函数
def f(x):
    return math.e**x

# 调用函数绘制y=x^2的图形
plot_function(f, 0, 100, title='Plot of y = x^2')
```
